### CLEAN THE DATA (remove unwanted columns- rename rows - make it easy to change out columns)

```python
#COVID CASES - df cases
#remove N/A zipcode data
cases = cases.dropna()
#convert to int 
cases = cases.astype({"Zip": int})
#convert to date
cases['Specimen Collection Date'] = pd.to_datetime(cases['Specimen Collection Date'], format='%m/%d/%Y')

sum_cases = cases.groupby(['Zip']).sum()
sum_cases.drop()
sum_cases
```

```python
#VACCINATIONS- df vacc
#remove N/A zipcode data
vacc = vacc.dropna()
#convert to int 
vacc = vacc.astype({"zipcode": int})
vacc = vacc.astype({"count_given": int})
vacc = vacc.astype({"dose_nbr": int})
#convert to date
vacc['vacc_date'] = pd.to_datetime(vacc['vacc_date'], format='%Y-%m-%d')
#replace shots/brand and convert to int
vacc['protocol'] = vacc['protocol'].replace('one-shot', 1)
vacc['protocol'] = vacc['protocol'].replace('two-shot', 2)
vacc = vacc.astype({"protocol": int})
#add column for completed vs incomplete vaccinations
vacc['incomplete vaccinations'] = vacc.protocol - vacc.dose_nbr
```
